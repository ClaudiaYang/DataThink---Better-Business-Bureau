If we want to know which businesses are more likely to be accredited, we should look at those who are already accredited. Find the correlation between some characteristics and those who are already accredited.

<font size = 3>
top 10 demographics of accredited businesses

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is stored in 'your_dataset.csv'
demographic_address = pd.read_csv('/Users/yangzhiyuan/Desktop/Data_Think/demographic_address.csv')

# Filter cases where AB is True
ab_true_df = demographic_address[demographic_address['AB?'] == True]

# Count occurrences of each demographic
demographic_counts = ab_true_df['Demographic'].str.split(', ').explode().value_counts()

# Get the top 10 demographics
top_10_demographics = demographic_counts.head(10)

# Visualize the result
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_demographics.values, y=top_10_demographics.index, palette='viridis')
plt.title('Top 10 Demographic Characteristics when AB is True')
plt.xlabel('Count')
plt.ylabel('Demographic Characteristic')
plt.show()
```

<font size = 3>
top 10 Cities of accredited businesses

```python
# Count occurrences of each demographic
city_counts = ab_true_df['City'].str.split(', ').explode().value_counts()

# Get the top 10 demographics
top_10_cities = city_counts.head(10)

# Visualize the result
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_cities.values, y=top_10_cities.index, palette='viridis')
plt.title('Top 10 Cities when AB is True')
plt.xlabel('Count')
plt.ylabel('Cities')
plt.show()
```

<font size = 3>
top 10 counties of accredited businesses

```python
# Count occurrences of each demographic
county_counts = ab_true_df['County'].str.split(', ').explode().value_counts()

# Get the top 10 demographics
top_10_counties = county_counts.head(10)

# Visualize the result
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_counties.values, y=top_10_counties.index, palette='viridis')
plt.title('Top 10 Counties when AB is True')
plt.xlabel('Count')
plt.ylabel('Counties')
plt.show()
```
